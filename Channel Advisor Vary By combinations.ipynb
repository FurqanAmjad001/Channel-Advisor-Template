```python
import itertools
from itertools import combinations
elements = ['Color', 'Size', 'Pattern', 'Style', 'Condition', 'Condition Grading', 'Memory', 'Model', 'Model Number', 'Pack Size', 'Storage','value']
combinations = []
for i in range(1,5):
    combinations += list(itertools.combinations(elements,i))
print(combinations)
```

```python
import pandas as pd
import numpy as np
```
